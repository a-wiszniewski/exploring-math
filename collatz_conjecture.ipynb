# The Collatz conjecture (the 3n + 1 problem)
All positive integers will reach 1 after a finite number of steps of the following arithmetic process:
* if the number is even, divide by 2
* if the number is odd, triple it and add 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# for a given integer, find the number of steps to 1 and the maximum integer calculated in the process

def collatz_conj(num=27):
    if num < 1:
        return None
    elif num == 1:
        return 0, 1
    else:    
        count = 0
        num_max = num
        while num != 1:
            count += 1
            if num % 2 == 0:
                num = num/2
            else:
                next_num = num*3 + 1
                if num_max < next_num:
                    num_max = next_num
                num = next_num    
        return count, int(num_max)
```

```python
# calculate number of steps to 1 and maximum integer calculated in the process, for a list of integers from 1 to n

max_int = 10**7

num_list = []
count_list = []
trajectory_list = []

for x in range(1, max_int):
    num_list.append(x)
    count_list.append(collatz_conj(x)[0])
    trajectory_list.append(collatz_conj(x)[1])
```

## Number of steps to reach 1 for a range of integers

### For integers up to 10<sup>5</sup>

```python
plt.figure(figsize=(8,4),dpi=200)
plt.plot(num_list,count_list, 'bo',markersize=0.01)
plt.xlabel('Input integer')
plt.ylabel('Number of steps to 1')
plt.xlim(0,10**5)
plt.ylim(0,350);
```

### For integers up to 10<sup>7</sup>

```python
plt.figure(figsize=(8,4),dpi=200)
plt.plot(num_list,count_list, 'bo',markersize=0.01)
plt.xlabel('Input integer')
plt.ylabel('Number of steps to 1');
```

## Distribution of number of steps to reach 1

```python
# calculate number of bins with FreedmanDiaconis rule

q25, q75 = np.percentile(count_list, [25, 75])
bin_width = 2 * (q75 - q25) * len(count_list) ** (-1/3)
bins = int(round((max(count_list) - min(count_list)) / bin_width))
```

```python
plt.figure(figsize=(7,3),dpi=200)
plt.hist(count_list, bins=bins)
plt.xlabel("Number of steps to 1")
plt.ylabel("Frequency");
```

## Following the trajectories of the integers as they reach 1, what is the maximum value they reach in the process
